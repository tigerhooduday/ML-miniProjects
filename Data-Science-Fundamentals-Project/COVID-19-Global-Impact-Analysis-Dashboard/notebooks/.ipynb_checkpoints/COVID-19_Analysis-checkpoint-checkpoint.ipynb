## Step 4: Reshape Time Series Data

melt() reshapes the DataFrame from wide to long format.

This makes plotting and analysis easier.

We convert 'Date' column from string to datetime using pd.to_datetime().

```python
# Convert confirmed data to long format
confirmed_long = confirmed.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    var_name='Date',
    value_name='Confirmed'
)
# Convert 'Date' column to datetime
confirmed_long['Date'] = pd.to_datetime(confirmed_long['Date'])

# Similarly for deaths
deaths_long = deaths.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    var_name='Date',
    value_name='Deaths'
)
# Convert 'Date' column to datetime
deaths_long['Date'] = pd.to_datetime(deaths_long['Date'])

# Similarly for recovered
recovered_long = recovered.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    var_name='Date',
    value_name='Recovered'
)
# Convert 'Date' column to datetime
recovered_long['Date'] = pd.to_datetime(recovered_long['Date'])



```

```python
confirmed.info()
confirmed_long.info()
```

## ✅ Step 5: Merge the Datasets

Combine all 3 datasets into one for easier analysis.

```python
# Merge confirmed, deaths, and recovered
full_data = confirmed_long.merge(deaths_long, on=['Province/State', 'Country/Region', 'Lat', 'Long', 'Date'])
full_data = full_data.merge(recovered_long, on=['Province/State', 'Country/Region', 'Lat', 'Long', 'Date'])
```

```python
full_data.info()
```

```python
full_data.tail()
```

## ✅ Step 6: Analyze Global Trends

1. Total Confirmed, Deaths, Recovered Worldwide

```python
global_daily = full_data.groupby('Date')[['Confirmed', 'Deaths', 'Recovered']].sum()
```

What’s happening:

    We group by Date and sum the cases to get total daily global numbers.

```python
# we use Matplotlib to plot them.
plt.figure(figsize=(12,6))
global_daily.plot()
plt.title('Global COVID-19 Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.grid(True)
plt.show()
```

## ✅ Step 7: Top 15 Countries by Total Confirmed Cases

```python
latest_date = full_data['Date'].max()

latest_data = full_data[full_data['Date'] == latest_date]
top_countries = latest_data.groupby('Country/Region')['Confirmed'].sum().sort_values(ascending=False).head(15)

# Plot
plt.figure(figsize=(15,6))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 15 Countries with Most Confirmed Cases')
plt.xlabel('Confirmed Cases')
plt.ylabel('Country')
plt.show() 
```

## ✅ Step 8: Death Rate and Recovery Rate

```python
latest_summary = latest_data.groupby('Country/Region')[['Confirmed', 'Deaths', 'Recovered']].sum()
latest_summary['Death Rate (%)'] = (latest_summary['Deaths'] / latest_summary['Confirmed']) * 100
latest_summary['Recovery Rate (%)'] = (latest_summary['Recovered'] / latest_summary['Confirmed']) * 100

# Show top 10 countries by death rate
latest_summary.sort_values(by='Death Rate (%)', ascending=False).head(10)
```

```python
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display
```

### ✅ STEP 3: Create a Country Dropdown Widget

```python
# Get sorted list of unique countries
countries = sorted(full_data['Country/Region'].unique())

# Create dropdown
country_dropdown = widgets.Dropdown(
    options=countries,
    value='India',  # default value
    description='Country:',
    style={'description_width': 'initial'}
)

display(country_dropdown)
```

### ✅ STEP 4: Interactive Line Chart of Confirmed, Deaths, Recovered

```python
def plot_country_trends(country):
    country_data = full_data[full_data['Country/Region'] == country]
    country_daily = country_data.groupby('Date')[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()

    fig = px.line(
        country_daily,
        x='Date',
        y=['Confirmed', 'Deaths', 'Recovered'],
        title=f'COVID-19 Trend in {country}',
        labels={'value': 'Cases', 'variable': 'Category'},
    )
    fig.update_layout(legend=dict(x=0, y=1))
    fig.show()

# Link function to dropdown
```

```python
widgets.interact(plot_country_trends, country=country_dropdown)
```

### ✅ STEP 5: Bar Chart — Top 10 Countries (Interactive Slider by Date)

```python
# Get unique dates
available_dates = sorted(full_data['Date'].unique())

# Date slider
date_slider = widgets.SelectionSlider(
    options=available_dates,
    value=available_dates[-1],
    description='Select Date:',
    layout={'width': '70%'},
    style={'description_width': 'initial'}
)

display(date_slider)

# Plot top 10 countries for that date
def plot_top10(date):
    day_data = full_data[full_data['Date'] == date]
    top10 = day_data.groupby('Country/Region')['Confirmed'].sum().sort_values(ascending=False).head(10).reset_index()

    fig = px.bar(
        top10,
        x='Confirmed',
        y='Country/Region',
        orientation='h',
        title=f'Top 10 Countries with Most Confirmed Cases on {date.date()}',
        labels={'Confirmed': 'Confirmed Cases', 'Country/Region': 'Country'},
    )
    fig.update_layout(yaxis={'categoryorder': 'total ascending'})
    fig.show()

widgets.interact(plot_top10, date=date_slider)


```

## Choropleth Map of COVID-19 Spread
-----

Step 1: Prepare the Data

We need to get the latest confirmed cases grouped by country.

```python
# Get latest date
latest_date = full_data['Date'].max()

# Filter data for the latest date
latest_data = full_data[full_data['Date'] == latest_date]

# Group by country and sum confirmed cases
country_cases = latest_data.groupby('Country/Region')['Confirmed'].sum().reset_index()
country_cases.columns = ['Country', 'TotalConfirmed']

# View top few rows
country_cases.head()
```

🌍 Step 2: Create the Choropleth Map

```python
import plotly.express as px

fig = px.choropleth(
    country_cases,
    locations='Country',
    locationmode='country names',  # can also use 'ISO-3' if available
    color='TotalConfirmed',
    hover_name='Country',
    color_continuous_scale='Reds',
    title=f'Global COVID-19 Confirmed Cases (as of {latest_date.date()})',
)

fig.update_geos(showframe=False, showcoastlines=False)
fig.update_layout(margin={"r":0,"t":50,"l":0,"b":0})
fig.show()
```

 Deaths or Recovered Map

```python
import plotly.express as px

# Prepare data
country_deaths = latest_data.groupby('Country/Region')['Deaths'].sum().reset_index()
country_deaths.columns = ['Country', 'TotalDeaths']

# Create map
fig = px.choropleth(
    country_deaths,
    locations='Country',
    locationmode='country names',
    color='TotalDeaths',
    hover_name='Country',
    color_continuous_scale='Reds',
    title=f'Global COVID-19 Deaths (as of {latest_date.date()})'
)

fig.update_geos(showframe=False, showcoastlines=False)
fig.update_layout(margin={"r":0,"t":50,"l":0,"b":0})
fig.show()
```

```python
# Prepare data
country_recovered = latest_data.groupby('Country/Region')['Recovered'].sum().reset_index()
country_recovered.columns = ['Country', 'TotalRecovered']

# Create map
fig = px.choropleth(
    country_recovered,
    locations='Country',
    locationmode='country names',
    color='TotalRecovered',
    hover_name='Country',
    color_continuous_scale='Greens',
    title=f'Global COVID-19 Recoveries (as of {latest_date.date()})'
)

fig.update_geos(showframe=False, showcoastlines=False)
fig.update_layout(margin={"r":0,"t":50,"l":0,"b":0})
fig.show()
```
